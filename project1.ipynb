#### 17. Build a dataset with every possible (UserID, ProductID) pair (cross join)

```python
#mannual way
l1=list(userDf['UserID'])
l2=list(productDf['ProductID'])

res = [[i, j] for i in l1 for j in l2]
o1=[]
o2=[]
for item in res:
    o1.append(item[0])
    o2.append(item[1])
dict1={'UserID':o1, 'ProductID':o2}
new_df=pd.DataFrame(data=dict1)
new_df
```

```python
# using pandas multi index

l1=list(userDf['UserID'])
l2=list(productDf['ProductID'])
data = pd.MultiIndex.from_product([l1,l2], names = ["UserID", "ProductID"])

pd.DataFrame(index = data).reset_index()
```

```python

# cols list if contains atleast one null value
missing_info= list(df23.columns[df23.isnull().any()])
missing_info
```

```python
#printing number of total missing values in a particular column
for col in missing_info:
    num_missing = df23[df23[col].isnull() == True].shape[0]
    print('number missing for column {}: {}'.format(col, num_missing))
```

```python
 # print(df23.shape[0]) number of rows
#finding percentage of missing values column-wise
for col in missing_info:
    percent_missing = df23[df23[col].isnull() == True].shape[0]/df23.shape[0] 
    print('percent missing for column {}: {}'.format( col, percent_missing))
```
